```python
# from langchain.document_loaders import TextLoader
# from langchain.text_splitter import CharacterTextSplitter
# from langchain.embeddings import HuggingFaceEmbeddings
# from langchain.vectorstores import Chroma
# # from langchain.document_loaders import PyPDFLoader
# from langchain.document_loaders import CSVLoader

# from more_itertools import chunked  # Helps with splitting lists into batches

# # Use Hugging Face model
# embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# text_splitter = CharacterTextSplitter(
#     separator="\n",
#     chunk_size=100,
#     chunk_overlap=0
# )

# loader = CSVLoader("cleaned_data.csv")
# docs = loader.load_and_split(
#     text_splitter=text_splitter
# )


# # Assuming `docs` is a list of documents
# docs_batches = list(chunked(docs, 1000))  # Split documents into smaller batches of 1000

# for batch in docs_batches:
#     db = Chroma.from_documents(
#         batch,
#         embedding=embeddings,
#         persist_directory="emb"
#     )

# db = Chroma.from_documents(
#     docs,
#     embedding=embeddings,
#     persist_directory="emb"
# )

# results = db.similarity_search(
#     " What is the most important in asset_skills,asset_model,asset_building,case_type,work_order_type,work_order_sub_type,category"
# )

# for result in results:
#     print("\n")
#     print(result.page_content)


from langchain.document_loaders import CSVLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from more_itertools import chunked  # Helps with splitting lists into batches

# Use Hugging Face model
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Split documents into smaller chunks
text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=100,
    chunk_overlap=0
)

# Load CSV file
loader = CSVLoader("cleaned_data.csv")
docs = loader.load_and_split(
    text_splitter=text_splitter
)

# Split documents into batches of 1000
docs_batches = list(chunked(docs, 1000))

# Create the Chroma store with an initial batch
db = None
persist_directory = "emb"

for batch in docs_batches:
    if db is None:
        # Initialize Chroma store for the first batch
        db = Chroma.from_documents(
            batch,
            embedding=embeddings,
            persist_directory=persist_directory
        )
    else:
        # Add more documents in batches to the existing Chroma store
        db.add_documents(batch)

# Perform similarity search
results = db.similarity_search(
    "What is the most important in asset_skills, asset_model, asset_building, case_type, work_order_type, work_order_sub_type, category"
)

# Print search results
for result in results:
    print("\n")
    print(result.page_content)

```

```python
from langchain.embeddings.base import Embeddings
from langchain.vectorstores import Chroma
from langchain.schema import BaseRetriever


class RedundantFilterRetriever(BaseRetriever):
    embeddings: Embeddings
    chroma: Chroma

    def get_relevant_documents(self, query):
        # calculate embeddings for the 'query' string
        emb = self.embeddings.embed_query(query)

        # take embeddings and feed them into that
        # max_marginal_relevance_search_by_vector
        return self.chroma.max_marginal_relevance_search_by_vector(
            embedding=emb,
            lambda_mult=0.8
        )

    async def aget_relevant_documents(self):
        return []
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
# from redundant_filter_retriever import RedundantFilterRetriever
import langchain

langchain.debug = True

from langchain.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_name = "EleutherAI/gpt-neo-1.3B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_length=1000,
    max_new_tokens=200 )
chat = HuggingFacePipeline(pipeline=pipe)

# Hugging Face embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = Chroma(
    persist_directory="emb",
    embedding_function=embeddings
)

retriever = RedundantFilterRetriever(
    embeddings=embeddings,
    chroma=db
)

chain = RetrievalQA.from_chain_type(
    llm=chat,
    retriever=retriever,
    chain_type="stuff"
)

result = chain.run("What is the most important in asset_skills, asset_model, asset_building, case_type, work_order_type, work_order_sub_type, category"
)
print(result)
```
